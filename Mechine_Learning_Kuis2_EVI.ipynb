```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt

def extract_text_from_ktp(image_path):
    try:
        # Baca gambar menggunakan Pillow (PIL)
        image = Image.open(image_path)

        # Gunakan Tesseract untuk mengenali teks dari gambar
        text = pytesseract.image_to_string(image)

        # Ganti tanda yang rancu
        text = text.replace('!', 'l')
        text = text.replace('=', ':')

        return image, text
    except Exception as e:
        return None, str(e)

# Contoh penggunaan
input_image_path = "/content/drive/MyDrive/Tugas Kuliah /Pembelajaran Mesin/KTP.jpg"
image, output_text = extract_text_from_ktp(input_image_path)

# Tampilkan gambar dengan teks OCR dalam 1 baris 2 kolom
if image is not None:
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    # Subplot pertama: Gambar Asli
    axs[0].imshow(image)
    axs[0].set_title('Gambar Asli')
    axs[0].axis('off')

    # Subplot kedua: Hasil Teks OCR
    axs[1].text(0.1, 0.5, output_text, verticalalignment='center', fontsize=12)
    axs[1].axis('off')

    plt.show()
```
